```python
import os
import zipfile

ZIP_PATH = "fer2013.zip"   # if you have this file locally, place it in repo root
EXTRACT_DIR = "dataset"    # this notebook expects final images under dataset/train/<class>/

if os.path.exists(ZIP_PATH):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_DIR)
    print(f"Extracted {ZIP_PATH} â†’ {EXTRACT_DIR}")
else:
    print("fer2013.zip not found. Please download FER2013 (Kaggle) and place fer2013.zip in the repo root.")
    print("Alternatively, prepare a folder structure: dataset/train/<class_name>/*.jpg")
```
